```python
import os
from dotenv import load_dotenv
load_dotenv() ## aloading all the environment variable

groq_api_key=os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
model=ChatGroq(model="llama-3.3-70b-versatile",groq_api_key=groq_api_key)
model
```

```python
from langchain_core.messages import HumanMessage
model.invoke([HumanMessage(content="Hi , My name is Pranathi and I am student")])
```

```python
from langchain_core.messages import AIMessage
model.invoke(
    [
        HumanMessage(content="Hi , My name is Pranathi and I am student"),
        AIMessage(content="Hello Pranathi! Nice to meet you! How's your day going so far? What are you studying, if you don't mind me asking?"),
        HumanMessage(content="Hey What's my name and what do I do?")
    ]
)
```

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

store={}

def get_session_history(session_id:str)->BaseChatMessageHistory:
    if session_id not in store:
        store[session_id]=ChatMessageHistory()
    return store[session_id]

with_message_history=RunnableWithMessageHistory(model,get_session_history,)
```

```python
config={"configurable":{"session_id":"chat1"}}
```

```python
with_message_history.invoke(
    [HumanMessage(content="Hi, My name is Pranathi and I am a Student")],
    config=config
)
```

```python
config1={"configurable":{"session_id":"chat2"}}
response=with_message_history.invoke(
    [HumanMessage(content="Whats my name")],
    config=config1
)
response.content
```

```python
response=with_message_history.invoke(
    [HumanMessage(content="hi,my name is ammu")],
    config=config1
)
response.content
```

```python
response=with_message_history.invoke(
    [HumanMessage(content="Whats my name")],
    config=config1
)
response.content
```

Prompt Templates

```python
from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder
prompt=ChatPromptTemplate.from_messages(
    [
        ("system","You are a helpful assistant.Amnswer all the question to the nest of your ability"),
        MessagesPlaceholder(variable_name="messages")
    ]
)
chain=prompt|model
```

```python
chain.invoke({"messages":[HumanMessage(content="Hi My name is Pranathi")]})
```

```python
with_message_history=RunnableWithMessageHistory(chain,get_session_history)
```

```python
config = {"configurable": {"session_id": "chat3"}}
response=with_message_history.invoke(
    [HumanMessage(content="Hi My name is Pranathi")],
    config=config
)
response
```

```python
response = with_message_history.invoke(
    [HumanMessage(content="What's my name?")],
    config=config,
)

response.content
```

```python
prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a helpful assistant. Answer all questions to the best of your ability in {language}.",
        ),
        MessagesPlaceholder(variable_name="messages"),
    ]
)
chain = prompt | model
```

```python
response=chain.invoke({"messages":[HumanMessage(content="Hi My name is Ammu")],"language":"Hindi"})
response.content
```

```python
response=chain.invoke({"messages":[HumanMessage(content="Hi My name is Ammu")],"language":"Telugu"})
response.content
```

```python
with_message_history=RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="messages"
)
```

```python
config={"configurable":{"session_id":"chat4"}}
response=with_message_history.invoke(
    {'messages':[HumanMessage(content="hi,I am Pranathi")],"language":"Telugu"},
    config=config
)
response.content
```

```python
response = with_message_history.invoke(
    {"messages": [HumanMessage(content="whats my name?")], "language": "Telugu"},
    config=config,
)
```

```python
response.text
```

Managing the Conversation History

```python
from langchain_core.messages import SystemMessage,trim_messages
trimmer=trim_messages(
    max_tokens=70,
    strategy="last",
    token_counter=model,
    include_system=True,
    allow_partial=False,
    start_on="human"
)
messages = [
    SystemMessage(content="you're a good assistant"),
    HumanMessage(content="hi! I'm pandu"),
    AIMessage(content="hi!"),
    HumanMessage(content="I like chocolate ice cream"),
    AIMessage(content="nice"),
    HumanMessage(content="whats 2 + 2"),
    AIMessage(content="4"),
    HumanMessage(content="thanks"),
    AIMessage(content="no problem!"),
    HumanMessage(content="having fun?"),
    AIMessage(content="yes!"),
]
trimmer.invoke(messages)
```

```python
from operator import itemgetter
from langchain_core.runnables import RunnablePassthrough
chain=(
    RunnablePassthrough.assign(messages=itemgetter("messages")|trimmer)
    |prompt
    |model
)
```

```python
response=chain.invoke(
    {
    "messages":messages+[HumanMessage(content="What ice cream do i like")],
    "language":"Telugu"
    }
)
response.content
```

```python
response=chain.invoke(
    {
    "messages":messages+[HumanMessage(content="What math problem did  ask")],
    "language":"English"
    }
)
response.content
```

```python
with_message_history = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="messages",
)
config={"configurable":{"session_id":"chat5"}}
```

```python
response=chain.invoke(
    {
    "messages":messages+[HumanMessage(content="What is my name")],
    "language":"English"
    },
    config=config,
)
response.content
```
